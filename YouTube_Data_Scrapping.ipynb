## Import Library

```python
# Import Library
import pandas as pd
import os
from googleapiclient.discovery import build
```

## Data Scrapper

```python
# Generate API YouTube V3
api_service_name = 'youtube'
api_version = 'v3'
youTubeApiKey = 'masukkan key yt mu'

# Memakai API Keys
youtube = build('youtube','v3', developerKey = youTubeApiKey)

# Request data dari video yang diinginkan
request = youtube.commentThreads().list(
    part = 'snippet',
    videoId = 'IX82qEPYyG0',
    maxResults = 1183
)

comments = []

while request is not None:
    response = request.execute()

    # Pengambilan komentar dari request
    for item in response['items']:
        comment = item['snippet']['topLevelComment']['snippet']
        public = item['snippet']['isPublic']
        comments.append([
            comment['authorDisplayName'],
            comment['publishedAt'],
            comment['likeCount'],
            comment['textOriginal'],
            public
        ])

    # Mengambil halaman berikutnya jika ada
    request = youtube.commentThreads().list_next(request, response)

# Membuat DataFrame dari komentar
df = pd.DataFrame(comments, columns=['author', 'date', 'like', 'comment', 'public'])
df.info()
```
